```python
def plot_box_sat(df, title, ls_columns, x_label, y_label, test_type=False, oreint = 'h'):
    plt.figure(figsize=(13,6))
    plt.title(title)
    
    # Create the boxplot with the corrected 'palette' argument
    ax = sns.boxplot(data=df[ls_columns], color  ='lightblue')
    
    # Set x and y labels
    plt.ylabel(y_label)
    plt.xlabel(x_label)
    
    # Show the plot
    plt.show()
    
```

```python
plot_box_sat(merge_sat, 'SAT participation in 2017-19', ['participation_17', 'participation_18', 'participation_19'], 
         'Year', 'Participation Rate')
plt.savefig('../images/SAT participation in 2017-19.png', format='png')
```
